```python
import os
import pandas as pd
import win32com.client as win32


def get_files_with_mod(folder_path):
    selected_files = []
    # List all files in the folder
    files = os.listdir(folder_path)
    # Filter out files containing "MOD" in their name
    for file_name in files:
        if "MOD" in file_name:
            selected_files.append(os.path.join(folder_path, file_name))
    return selected_files




def send_email_with_attachments(subject, body, recipient, attachments, cc, bcc, shared_mailbox):
    outlook = win32.Dispatch('Outlook.Application')
    namespace = outlook.GetNamespace("MAPI")
    
    # Get the shared mailbox account
    recipient = namespace.CreateRecipient(shared_mailbox)
    recipient.Resolve()
    mail = outlook.CreateItem(0)# 0 represents an email item
    mail.SentOnBehalfOfName = shared_mailbox
    mail.Subject = subject
    mail.Body = body
    mail.To = recipient

    if cc:
        mail.CC = cc

    if bcc:
        mail.BCC = bcc

    # Attach all files
    for attachment_path in attachments:
        mail.Attachments.Add(attachment_path)

    mail.Send()



excel_file_path ='C://Users//SJA451//OneDrive - Maersk Group//Documents//ASCII Characters Replacement//ASCII Character Replacement data.xlsx'
df = pd.read_excel(excel_file_path, sheet_name='Sheet1')

for index, row in df.iterrows():
    incident_number = row['Incident']
    mail_status = row['Email Sent']
    short_description=row['Short Description']
    
    # Check if mail status is empty
    if mail_status!= "Email Sent":
        folder_path = f'C:/Users/SJA451/OneDrive - Maersk Group/Documents/ASCII Characters Replacement/{incident_number}'  
        files_with_mod = get_files_with_mod(folder_path)
        # List files in the folder containing "MOD" in their name
        
        
        if files_with_mod:
            subject =f'Incident Number {incident_number} - {short_description}'
            body = " Hello OT Team,\n Please find the attached corrected file for reprocessing. Do revert with your feedback.\n Thanks & Regards,\n Automation and Enablement Team"
            recipient = "lakshay.jain@maersk.com"
            cc = None
            bcc = None
            shared_mailbox = "automationandenablement@maersk.com" 

            send_email_with_attachments(subject, body, recipient, files_with_mod, cc, bcc, shared_mailbox)
            print("Email sent successfully with files containing 'MOD' in their name.")
        else:
            print("No files with 'MOD' in their name found in the folder.")
            
          
```
